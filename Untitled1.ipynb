```python
# ==============================================================================
# PROJECT: Fall Risk Factors in Older Adults (ELSI-Brasil)
# AUTHOR: [Seu Nome]
# GOAL: Multivariate Logistic Regression for Digital Science Master's preparation
# DATA SOURCE: ELSI-Brasil (English Longitudinal Study of Ageing)
# ==============================================================================

import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Carregamento com tratamento de erro
try:
    df = pd.read_stata('elsi_data.dta')

    # 2. Seleção de variáveis e limpeza
    # d16: Queda | d1: HAS | sexo: Gênero | idade: Idade
    cols = ['d16', 'sexo', 'idade', 'd1']
    df_clean = df[cols].copy()

    # Transformando em binário (0 e 1) e limpando "Não sabe/Não respondeu"
    for col in ['d16', 'sexo', 'd1']:
        df_clean[f'{col}_code'] = df_clean[col].cat.codes
        df_clean = df_clean[df_clean[f'{col}_code'].isin([0, 1])]

    df_clean['idade_num'] = pd.to_numeric(df_clean['idade'], errors='coerce')
    df_clean = df_clean.dropna()

    print(f"✅ Limpeza concluída! Analisando {len(df_clean)} pacientes.")
except FileNotFoundError:
    print("❌ Erro.")
```

```python
# ==============================================================================
# TABLE 1: SAMPLE CHARACTERISTICS (DESCRIPTIVE STATISTICS)
# ==============================================================================

# Creating a nice descriptive table
table1 = pd.DataFrame()

# Gender distribution
gender_counts = df_clean['sexo'].value_counts(normalize=True) * 100
table1.loc['Gender: Female (%)', 'Full Sample'] = f"{gender_counts['Feminino']:.1f}%"

# Hypertension prevalence
has_counts = df_clean['d1'].value_counts(normalize=True) * 100
table1.loc['Hypertension: Yes (%)', 'Full Sample'] = f"{has_counts['Sim']:.1f}%"

# Age statistics
table1.loc['Age: Mean (SD)', 'Full Sample'] = f"{df_clean['idade_num'].mean():.1f} (±{df_clean['idade_num'].std():.1f})"

# Target variable: Falls
fall_counts = df_clean['d16'].value_counts(normalize=True) * 100
table1.loc['History of Falls: Yes (%)', 'Full Sample'] = f"{fall_counts['Sim']:.1f}%"

print("--- TABLE 1: SAMPLE CHARACTERISTICS (n=9,699) ---")
print(table1)
print("-" * 50)
```

```python
# ==============================================================================
# 3. ADJUSTED MULTIVARIATE MODEL & PROFESSIONAL VISUALIZATION
# ==============================================================================

# A. Running the Model
X = sm.add_constant(df_clean[['idade_num', 'sexo_code', 'd1_code']])
y = df_clean['d16_code']
model = sm.Logit(y, X).fit()

# B. Preparing Results Table
results = pd.DataFrame({
    'OR': np.exp(model.params),
    'P': model.pvalues,
    'Lower': np.exp(model.conf_int()[0]),
    'Upper': np.exp(model.conf_int()[1])
})
results.index = ['Intercept', 'Age (Years)', 'Gender (Female)', 'Hypertension (HAS)']
final_results = results.drop('Intercept')

# C. Professional Plotting
plt.figure(figsize=(12, 7))
sns.set_style("white")

# Data for plotting
plot_data = final_results.reset_index()
plot_data.columns = ['Factor', 'OR', 'P', 'Lower', 'Upper']

# Creating the error bars (Confidence Intervals)
for i in range(len(plot_data)):
    plt.hlines(y=plot_data['Factor'][i], xmin=plot_data['Lower'][i],
               xmax=plot_data['Upper'][i], color='black', alpha=0.5, linewidth=2)

# Plotting the Odds Ratio points
# Red for significant (p < 0.05), Grey for non-significant
colors = ['#d62728' if p < 0.05 else '#7f8c8d' for p in plot_data['P']]
sns.scatterplot(x='OR', y='Factor', data=plot_data, s=200, hue='Factor',
                palette=colors, legend=False, zorder=3)

# Add vertical line at OR = 1.0 (Neutral Risk)
plt.axvline(x=1, color='#c0392b', linestyle='--', linewidth=1.5, label='Neutral Risk (OR=1.0)')

# Refning titles and labels (Academic English)
plt.title('Independent Risk Factors for Falls in Older Adults\n(ELSI-Brasil Multivariate Analysis)',
          fontsize=16, fontweight='bold', pad=20, loc='left')
plt.xlabel('Odds Ratio (95% Confidence Interval)', fontsize=12, fontweight='bold')
plt.ylabel('')
plt.xlim(0.8, 2.0) # Adjusting scale for better visibility

# Adding labels for the P-values and OR values
for i, row in plot_data.iterrows():
    plt.text(row['OR'], i + 0.15, f"OR: {row['OR']:.2f} (p={row['P']:.3f})",
             ha='center', fontsize=10, fontweight='bold', color='#2c3e50')

sns.despine() # Removes top and right borders for a modern look
plt.tight_layout()
plt.show()

# Final Summary Table
print("\n" + "="*50)
print("FINAL SUMMARY TABLE (ACADEMIC FORMAT)")
print("="*50)
print(final_results.sort_values(by='OR', ascending=False))
```

```python
# ==============================================================================
# MODEL DIAGNOSTICS
# ==============================================================================

print("\n--- MODEL PERFORMANCE METRICS ---")
# Pseudo R-Squared (McFadden's)
print(f"Pseudo R-Squared (McFadden): {model.prsquared:.4f}")
# Log-Likelihood
print(f"Log-Likelihood: {model.llf:.2f}")
# LLR p-value (Model Significance)
print(f"Model LLR p-value: {model.llr_pvalue:.2e}")

print("\nClinical Note: A significant LLR p-value (< 0.05) indicates that our ")
print("independent variables significantly improve the model fit compared to a null model.")
```
